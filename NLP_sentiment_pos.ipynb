```python
#Remove ads in spanish
sex_ads = sex_ads[sex_ads['spanish_g'] != 1]
```

```python
#Now time to use an NLP pipeline!
sex_ads_docs = list(nlp.pipe(sex_ads['title_label']))
```

```python
#I then wrangle my dictonary using the list comphresion approach we learned in class
os.chdir("/content/drive/MyDrive/Colab Notebooks/Data/sentiment_datasets")

nrc_df = pd.read_csv('nrc.csv')

nrc_dict = {col.split('_')[0].lower(): nrc_df[col].dropna().tolist() for col in nrc_df.columns}

print(nrc_dict["negative"][0:11]) #first 12 values for negative
print(nrc_dict["positive"][-11:]) #the last 12 values for positive
```

```python
#Create my safe divide function for later (this is also from class)

def safe_divide(a, b):
    if b != 0:
        return a/b
    else:
        return 0
```

```python
#Now, I am going to use the for loop techinque we learned in class to determine the number of positive and negative words, and the % of positive and negative words
#In each ad
# Extract terms with emotion
final_nw = []
final_positive = []
final_negative = []

for doc in sex_ads_docs: #go through each document
  negative_count = 0
  positive_count = 0
  words = 0
  for token in doc: #go through each token in the document
    if not token.is_punct and not token.is_space: # Removed the check for is_emoji
      words += 1
      if token.lower_ in nrc_dict["negative"]:
        negative_count += 1
      if token.lower_ in nrc_dict["positive"]:
        positive_count += 1
  final_positive.append(safe_divide(positive_count, words)) #divide by number of words that have concreteness scores
  final_negative.append(safe_divide(negative_count, words))
  final_nw.append(words)


sex_ads['NW'] = final_nw #create a new column in pandas called NW and assign it the values from final_nw
sex_ads['NRC_positive'] = final_positive
sex_ads['NRC_negative'] = final_negative
```

```python

nw_final = [] # word count
verb_final = []

#loop for each document in spacy doc
for doc in sex_ads_docs:
    nw = 0 #starting count for number of words
    verb = 0

    for token in doc: #for token in each doc
        if token.is_punct or token.is_space:
          continue
        nw += 1
        if token.pos_ == "VERB":
            verb += 1

    #print(nw) #this will print by text
    nw_final.append(nw) #append number of words by text to list above (creating a list of ints)
    verb_final.append(safe_divide(verb, nw))
```

```python
sex_ads['NW'] = nw_final #create a new column in pandas called NW and assign
sex_ads['verb_final'] = verb_final
# Export the DataFrame to a CSV file
sex_ads.to_csv('sex_ad_w_nlp_final.csv', index=False)
```
